```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
# Draw a distplot
sns.distplot(penguins.flipper_length_mm, rug=True)
```

```python
# Removing the KDE plot
sns.distplot(penguins.flipper_length_mm, hist=False)
```

```python
# Setting the number of bins
sns.distplot(penguins.flipper_length_mm, bins=20)
```

```python
# Bin values as an array
sns.distplot(penguins.flipper_length_mm, bins=[160, 170, 180, 190, 200, 240])
```

```python
# Fitting the data to a particular distribution using norm_list
sns.distplot(penguins.flipper_length_mm, norm_hist=True, kde =False)
```

```python
# Styling Rugplots
sns.distplot(penguins.flipper_length_mm, rug_kws= {'color':'red', 'height':0.30}, rug=True)
```

```python
# Styling Histograms
sns.distplot(penguins.flipper_length_mm, kde = False, hist_kws={'color':'blue', 'alpha':0.3, 'histtype':'stepfilled'})
```

```python
# STyling KDE Plots
sns.distplot(penguins.flipper_length_mm, kde=True, hist= False, kde_kws= {'color':'red', 'lw':2, 'alpha':0.4})
```
