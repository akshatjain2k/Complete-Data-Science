```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
df = pd.read_csv("lrsv.csv")
df
```

```python
plt.xlabel('area(squr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area, df.price, color='red', marker ='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)    # training model using the availble datapoints
```

```python
reg.predict([[3300]])
```

```python
reg.coef_            # value of m in y = mx + c
```

```python
reg.intercept_        # value of b in y = mx + c
```

```python
# y = m*x  + b
y = 135.78767123*3300 + 180616.43835616432
y                                            ## same as predicted
```

```python
plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area, df.price, color='red', marker ='+')
plt.plot(df.area, reg.predict(df[['area']]), color = 'blue')
```

```python
d = pd.read_csv("areas.csv")
```

```python
p = reg.predict(d)
```

```python
d['prices'] = p
```
