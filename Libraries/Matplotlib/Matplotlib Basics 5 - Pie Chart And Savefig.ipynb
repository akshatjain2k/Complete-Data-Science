```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_val = [1400, 600, 300, 410, 250]
exp_label = ["Home Rent", "Food", "Phone/Internet Bil", "Car", "Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_val, labels=exp_label, radius=1.5, autopct='%0.1f%%', shadow=True, explode=[0,0,0.25,0,0], startangle=45)
plt.show()
```

```python
plt.savefig("piecharty.png",bbox_inches="tight", transparent=True)
```
